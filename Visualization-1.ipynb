```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.array([1,2,3,4,5,6])
y=np.power(x,3)
plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.title('My First Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.style.use('classic')
plt.show()
```

```python
x=np.linspace(0,10,1000)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

# Seaborn

* Categorical Plot  ---- At least one categorical Column
 1   countplot 
 2  boxplot
 3 barplot
 4  violinplot
 5   swarm plot
* Distribution Plot ---- At least one cont numerical Column
    distplot
 1  histogram
 2 kde plot
 3  line plot
 4   regplot
 5  jointplot
 
* Matrix            ---- Cont, num, categorical()
 1   pairplot
 2  heatmap

* Regression Plot   ---- Two cont column
 1 scatter plot
 2 regplot
